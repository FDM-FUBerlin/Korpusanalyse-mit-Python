# Pandas

```python
import pandas as pd
from os import listdir
```

```python
inputDir = "Daten_EL/Dataframes/Deixis/"
a = listdir(inputDir)
a
```

```python
df1 = pd.read_csv(inputDir + 'Erbauungsbuch.csv', sep=';', na_values=".")
df1
```

```python
df2 = pd.read_csv(inputDir + 'Leichenpredigten.csv', sep=';', na_values=".")
df2
```

```python
df3 = pd.read_csv(inputDir + 'Gebetbuch.csv', sep=';', na_values=".")
df3
```

```python
import matplotlib.pyplot as plt
```

```python
feat = '1tePerson'
```

```python
fig = plt.boxplot([df1[feat], df2[feat], df3[feat]], 
                  showmeans=True, 
                  showfliers=True, 
                  notch=1, 
                  medianprops={'color': 'black'}, 
                  meanprops=dict(markerfacecolor='None', markeredgecolor='black'), 
                  patch_artist=True)

fig['boxes'][0].set_facecolor('#a9cce3')
fig['boxes'][1].set_facecolor('#d6eaf8')
fig['boxes'][2].set_facecolor('#edbb99')

plt.title('Merkmal: ' + str(feat), fontsize=14)
plt.xticks([1, 2, 3], ['Erbauungsbuch', 'Leichenpredigten', 'Gebetbuch'], fontsize=12, rotation=15)
plt.yticks(fontsize=12)
plt.tight_layout()
```
